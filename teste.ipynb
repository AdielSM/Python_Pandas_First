```python
import pandas as pd
```

```python
document = pd.read_csv('large_dataset.csv')
#retorna as primeiras 5 linhas por default
document.head()
```

```python
document = document.rename(columns={"country": "País", "continent": "Continente", "year": "Ano", "lifeExp": "Expectativa de vida", "pop": "População", "gdpPercap": "PIB"})
document.head(10)
```

```python
#informações estatísticas sobre os dados
document.describe()
```

```python
#retorna quais os valores de continentes na base de dados
document['Continente'].unique()
```

```python
#função de filtro que expera uma expressão
americas = document.loc[document["Continente"] == 'Americas']
americas.head()
```

```python
#group by padrão (nunique = números, unique = dado)
document.groupby("Continente")["País"].nunique()
```

```python
#retornar a expectativa de vida média para cada ano
document.groupby("Ano")["Expectativa de vida"].mean()
```
